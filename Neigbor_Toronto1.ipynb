### Scrape the following Wikipedia page: https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M, and transform the data into a pandas dataframe

```python
import requests
url_wiki = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(url_wiki).text
```

```python
!pip install msgpack
!pip install beautifulsoup4
!pip install lxml
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(source, 'lxml')
table = soup.find('table', class_="wikitable sortable").tbody
tr1 = table.find('tr')
```

```python
column_names = list()
for th in tr1.find_all('th'):
    column_names.append(th.text.strip())
column_names
```

```python
rows = list()
for tr in table.find_all('tr'):
    row = list()
    for td in tr.find_all('td'):
        row.append(td.text.strip())
    rows.append(row)
rows = rows[1:]
rows
```

```python
import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import numpy as np # library to handle data in a vectorized manner
```

```python
neighborhoods = pd.DataFrame(rows, columns=column_names)
neighborhoods.head()
```

```python
neighborhoods = neighborhoods[neighborhoods['Borough']!='Not assigned'].reset_index(drop=True)
neighborhoods.head()
```

```python
neighbor = neighborhoods.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join) #combine rows of string within the same group
neighbor = neighbor.reset_index()
neighbor.head()
```

```python
neighbor.loc[neighbor['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = neighbor['Borough']
neighbor
```

```python
neighbor.to_csv('neighbor.csv', index = False)
```
